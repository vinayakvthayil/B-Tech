```python
#part A
import numpy as np
import matplotlib.pyplot as plt

# 1
# Create an array x and calculate the function h(x)
x_task1 = np.array([1, 1, 2, 3, 4, 3, 4, 6, 4])
t0 = 1.2
t1 = 0.5
h_x = t0 + t1 * x_task1

# Plot the graph
plt.figure(figsize=(8, 6))
plt.scatter(x_task1, h_x, color='blue', label='h(x) = 1.2 + 0.5*x')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.title('Linear Regression: h(x) = 1.2 + 0.5*x')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#4

A = np.array([1, 1, 2, 3, 4, 3, 4, 6, 4])
B = np.array([2, 1, 0.5, 1, 3, 3, 2, 5, 4])

n = len(A)
mse = np.sum((A - B) ** 2) / n

print(f"Mean Squared Error (MSE): {mse}")
```

```python
#5
def compute_cost_function(n, t1, A, B):
    h_x = t1 * A
    mse = np.sum((h_x - B) ** 2) / (2 * n)
    return mse

A = np.array([1, 1, 2, 3, 4, 3, 4, 6, 4])
B = np.array([2, 1, 0.5, 1, 3, 3, 2, 5, 4])

n = len(A)

# Iterate over different values of t1 and calculate the cost function
t1_values = np.linspace(-2, 2, 100)  # Adjust the range as needed
cost_values = []

for t1 in t1_values:
    cost = compute_cost_function(n, t1, A, B)
    cost_values.append(cost)

# Find the index of the minimum cost
min_cost_index = np.argmin(cost_values)
best_t1 = t1_values[min_cost_index]
min_cost = cost_values[min_cost_index]

# Print the results
print(f"Best t1: {best_t1}")
print(f"Minimum Cost: {min_cost}")

# Plot the cost function
import matplotlib.pyplot as plt

plt.plot(t1_values, cost_values, label='Cost Function')
plt.xlabel('t1')
plt.ylabel('Cost')
plt.title('Cost Function vs t1')
plt.legend()
plt.grid(True)
plt.show()
```
