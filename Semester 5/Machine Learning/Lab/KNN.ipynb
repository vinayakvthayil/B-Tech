```python
#9
M = np.random.rand(10,2)
M_dash = np.random.rand(5,2)
dist = np.zeros((10,5,3))
for i in range(10):
    for j in range(5):
        distance = np.linalg.norm(M[i] - M_dash[j])
        dist[i, j, 0] = i
        dist[i, j, 1] = j
        dist[i, j, 2] = distance
print(dist)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
distances_sklearn = euclidean_distances(M, M_dash)
print(distances_sklearn)
```

```python
#10
sorted_dist = dist[dist[:, 2].argsort()]
print(sorted_dist)
```

```python
#11
k=2
init = sorted_dist[: k] 
print(init)
```

```python
#12
ones = (init[:, 2] == 1).sum()
zeros = (init[:, 2] == 0).sum()
if ones > zeros:
    result = 1
else:
    result = 0
print(result)
```
